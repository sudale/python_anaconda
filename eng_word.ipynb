```python
import pandas as pd
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from wordcloud import STOPWORDS, WordCloud
import nltk
```

```python
# nltk.download()
```

```python
# riss_bigdata.csv 에서 제목 읽어서 출력
all_data = pd.read_csv('riss_bigdata.csv', encoding='utf-8')
all_data
```

```python
all_title = all_data['제목']
all_title
```

```python
stopWords = set(stopwords.words('english'))
lemma = WordNetLemmatizer()
stopWords
```

```python
first_title = all_title[0:1]
first_title
```

```python
EnWords = re.sub(r'[^a-zA-Z]+',' ',str(first_title))
EnWords
```

```python
EnWordsToken = word_tokenize(EnWords.lower())
EnWordsToken
```

```python
EnWordsTokenStop = [w for w in EnWordsToken if w not in stopWords]
EnWordsTokenStop
```

```python
EnWordsTokenStopLemma = [lemma.lemmatize(w) for w in EnWordsTokenStop]
EnWordsTokenStopLemma
```

```python
lemma = WordNetLemmatizer() # 표제어 추출 객체
words = []

for title in all_title:
    EnWords = re.sub(r'[^a-zA-Z]+',' ',str(title)) # 영어 아닌 것을 ' ' 대체 (^ not)
    EnWordsToken = word_tokenize(EnWords.lower()) # 소문자로
    EnWordsTokenStop = [w for w in EnWordsToken if w not in stopWords]
    EnWordsTokenStopLemma = [lemma.lemmatize(w) for w in EnWordsTokenStop] # 기본형태
    words.append(EnWordsTokenStopLemma)
    
words
```

```python
# 전처리가 끝난 words 를 1차원 list
from functools import reduce
words2= list(reduce(lambda x, y : x+y, words))
words2
```

```python
# 단어 빈도수 구하기
from collections import Counter
count = Counter(words2)
count
```

```python
# 상위 50개 단어 딕셔너리에 저장
word_count = dict()
# count.most_common(50)
```

```python
for tag, counts in count.most_common(50):
    if(len(str(tag))>5):
        word_count[tag] = counts
word_count
```

```python
for tag, counts in count.most_common(50):
    if(len(str(tag))>1):
        word_count[tag] = counts
        print("%s : %d" % (tag, counts))
```

```python
import matplotlib.pyplot as plt

# 히스토그램 그리기
plt.figure(figsize=(12,5))
plt.xlabel("word")
plt.ylabel("count")
plt.grid(True)

sorted_Keys = sorted(word_count, key=word_count.get, reverse=True)
sorted_Values = sorted(word_count.values(), reverse=True)

plt.bar(range(len(word_count)), sorted_Values, align='center')
plt.xticks(range(len(word_count)), list(sorted_Keys), rotation='85')

plt.show()
```
